```python
L1 = [1, 2, 3]
L2 = [1, 2, 3, 4, 5]

set1 = set(L1)
set2 = set(L2)

is_subset = set1.issubset(set2)

if is_subset:
    print("L1 is a subset of L2")
else:
    print("L1 is not a subset of L2")
```

```python
l1 = [1,2,50]
l2 = [1,2,3,4,5]

for i in l1:
    if i not in l2:
        print("Not a subset")
        break

    else:
        print("It is a subset")
```

```python
list1= [1,2,3,4,5]
list2= [1,2,3,5]
m=0
for i in range(len(list2)):
  if list2[i] in list1:
    m=1
  else:
    m=0
if m==1:
  print("subset")
else:
  print("not subset")
```
